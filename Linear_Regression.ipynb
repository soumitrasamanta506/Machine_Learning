```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day48-simple-linear-regression/placement.csv'
df = pd.read_csv(url)
```

```python
df.head(10)
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel('Placement(in LPA)')
```

```python
x = df.iloc[:, 0:1]
y = df.iloc[:, 1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
x_test
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.plot(x_train, lr.predict(x_train), color='red')
plt.xlabel('cgpa')
plt.ylabel('package')
```
